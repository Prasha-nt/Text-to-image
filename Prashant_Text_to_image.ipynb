```python
import torch
import gradio as gr
from diffusers import StableDiffusionPipeline

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")

# Move to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe.to(device)

# Function to generate an image from a text prompt
def generate_image(prompt):
    image = pipe(prompt, guidance_scale=7.5, num_inference_steps=50).images[0]
    return image

# Create a Gradio UI
interface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(placeholder="Enter a text prompt..."),
    outputs=gr.Image(),
    title="Stable Diffusion - Text to Image Generator",
    description="Enter a text prompt and generate an AI-powered image!"
)

# Launch the UI
interface.launch()
```
